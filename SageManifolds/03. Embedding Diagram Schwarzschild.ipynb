# 03. Embedding Diagram for the Schwarzschild Spacetime

### Abstract

In this notebook we present the embedding diagram for the Schwarzschild spacetime.

---


We begin our analysis by writting the 3-dimensional flat space in cylindrical coordinates,

\begin{equation}
ds^2 = dz^2 + dr^2 + r^2 d\phi^2
\end{equation}

and considering a (2-dimensional) surface given by the equation

\begin{equation} 
z= z(r).
\end{equation}

Hence, we have

\begin{equation} 
dz= \frac{dz}{dr}dr
\end{equation}

and therefore

\begin{equation}
ds^2 = \left[ 1+ \left(\frac{dz}{dr}\right)^2\right]dr^2 + r^2 d\phi^2.
\end{equation}

Comapring this relation with the Schwarzschild line element at a fixed time, $dt=0$, and for a fixed latitude,  $d\theta =0$,

\begin{equation}
ds^2 = \left( 1- \frac{2M}{r}\right)^{-1}dr^2 + r^2 d\phi^2,
\end{equation}

we can identify that

\begin{equation}
 1+ \left(\frac{dz}{dr}\right)^2 = \left( 1- \frac{2M}{r}\right)^{-1}.
\end{equation}

From this relation we obtain

\begin{align}
\left(\frac{dz}{dr}\right)^2 =& \left( 1- \frac{2M}{r}\right)^{-1} - 1 \\
\left(\frac{dz}{dr}\right)^2 =& \frac{r}{r-2M} - 1\\
\left(\frac{dz}{dr}\right)^2 =& \frac{2M}{r-2M} \\
\frac{dz}{dr} =& \sqrt{\frac{2M}{r-2M}}\\
dz =& \sqrt{2M} \frac{dr}{\sqrt{r-2M}}.
\end{align}

Integrating this expression we get
\begin{equation}
z(r) = \sqrt{8M}\sqrt{r-2M} .
\end{equation}

## Plotting using  `matplotlib.pyplot`

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import warnings
warnings.simplefilter('ignore')

# Mass of the central object 
M=5.

# Surface definition
x = np.arange(-100, 100, 0.25)
y = np.arange(-100, 100, 0.25)
x, y = np.meshgrid(x, y)

r = np.sqrt(x**2 + y**2)
z = np.sqrt(8*M)*np.sqrt(r - 2*M)

# 3D-Plot environment
fig, ax = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=(15,5))

# Plot the wireframe 
ax[0].plot_wireframe(x, y, z, rstride=20, cstride=20)

# Plot the surface.
ax[1].plot_surface(x, y, z,  rstride=20, cstride=20)

plt.show()
```

## Plot using `EinsteinPy`

```python
from einsteinpy.hypersurface import SchwarzschildEmbedding
from einsteinpy.plotting import HypersurfacePlotter
from astropy import units as u
```

```python
surface_obj = SchwarzschildEmbedding(5. * u.M_sun)
surface = HypersurfacePlotter(embedding=surface_obj, plot_type='surface')
surface.plot()
surface.show()
```
