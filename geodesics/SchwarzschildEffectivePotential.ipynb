### Summary

We show the behavior of the effective potential for a test particle moving around a Schwarzschild black hole.

---

```python
import matplotlib.pyplot as plt
import numpy as np

M = 1
delta = 0

def Veff(r,l):
    return (1 - 2*M/r)*(l**2/r**2 + delta)

def VeffN(r,l):
    return - M/r + l**2/r**2 + delta


rr = np.linspace(2,50,250)
l = 5


plt.figure()
plt.plot(rr, Veff(rr,5), color= 'crimson', label='Schwarzschild')
plt.plot(rr, Veff(rr,4), color= 'lime', label='Schwarzschild')
plt.plot(rr, Veff(rr,3), color= 'magenta', label='Schwarzschild')

#plt.plot(rr, VeffN(rr,l), color= 'cornflowerblue', label='Newton')
plt.axhline(0, color='black', alpha=0.3)
plt.axhline(1, color='black', alpha=0.3)
plt.axvline(3, color='black', alpha=0.7)
plt.xlabel(r'$r$')
plt.ylabel(r'$V_{eff} (r)$')
plt.grid()
plt.legend()
plt.show()
```
