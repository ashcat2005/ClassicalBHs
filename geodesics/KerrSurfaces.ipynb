### Summary

We show the important surfaces such as event horizon, photon sphere and ISCO around a Kerr black hole.

---

```python
import matplotlib.pyplot as plt
import numpy as np

M = 1

def event_horizon(a):
    return M + np.sqrt(M**2 - a**2)

def inner_horizon(a):
    return M - np.sqrt(M**2 - a**2)


def photon_sphere(a):
    co_rot = 2*M*(1 + np.cos(2/3*np.arccos(-a/M)))
    counter_rot = 2*M*(1 + np.cos(2/3*np.arccos(a/M)))
    return co_rot, counter_rot


def isco(a):
    z1 = M + (M**2 - a**2)**(1/3)*((M+a)**(1/3) + (M-a)**(1/3))
    z2 = np.sqrt(3*a**2 + z1**2)
    r_co_rot = 3*M + z2 - np.sqrt((3*M - z1)*(3*M + z1 + 2*z2))
    r_counter_rot = 3*M + z2 + np.sqrt((3*M - z1)*(3*M + z1 + 2*z2))
    return r_co_rot, r_counter_rot
```

```python
a_range = np.linspace(0, 1, 500)

eh = event_horizon(a_range)
ih = inner_horizon(a_range)
ps_co_rot, ps_counter_rot = photon_sphere(a_range)
isco_co_rot, isco_counter_rot = isco(a_range)

plt.figure()
plt.plot(a_range, eh, color= 'crimson', label='Event Horizon')
plt.plot(a_range, ih, '--', color= 'crimson',  label='Event Horizon')
plt.plot(a_range, ps_co_rot, color= 'cornflowerblue', label='Photon Sphere (corot.)')
plt.plot(a_range, ps_counter_rot, '--', color= 'cornflowerblue', label='Photon Sphere (counterrot.)')
plt.plot(a_range, isco_co_rot, color= 'mediumslateblue', label='ISCO (corot.)')
plt.plot(a_range, isco_counter_rot, '--', color= 'mediumslateblue', label='ISCO (counterrot.)')



plt.axhline(0, color='black', alpha=0.5)
plt.axvline(0, color='black', alpha=0.5)
plt.xlabel(r'$a$')
plt.ylabel(r'$r$')
plt.grid(alpha=0.2)
plt.legend()
plt.show()
```
